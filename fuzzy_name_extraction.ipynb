```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/parcel_grouped.csv')
```

```python
df['1748Owner']
```

```python
list_family_names = pd.read_csv('data/venice_families_cleaned.csv')
family_name_options = list_family_names['family'].values
```

```python
# we will use a fuzzy ratio to extract the name
from fuzzywuzzy import fuzz, process

top_3_names = df['1748Owner'].apply(lambda x: process.extract(str(x),family_name_options, scorer=fuzz.token_set_ratio, limit = 3))
```

```python
top_3_names
```

```python
owner_3 = ['owner_name1','owner_name2','owner_name3']
df[owner_3] = pd.DataFrame(top_3_names)['1748Owner'].apply(lambda x: pd.Series(x))
```

```python
for col in owner_3:
    df[col] = df[col].apply(lambda x: x[0] if x[1] > 95 else None)
    df[col] = df[col].apply(lambda x: x if x != 'di' and x != 'de' and x != 'della' else None)
```
