```python
import pandas as pd
import numpy as np 
```

```python
cadaster = pd.read_csv("data/cadaster_1808.csv")
```

```python
# check if cadaster contains NA
# we find that there are 737 rows where the parcel owner text is None
cadaster[cadaster['parcelOwnerText'].isna()].head(4)
```

```python
# we will drop those rows 
cadaster_nn = cadaster.dropna(subset=['parcelOwnerText'])
```

```python
# what words appear in the parcel owner text?
import matplotlib.pyplot as plt 
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(min_df=20)
X = vectorizer.fit_transform(cadaster_nn['parcelOwnerText'])
X = X.toarray()
word_counts = pd.DataFrame(X, columns = vectorizer.get_feature_names())
```

```python
word_counts.sum(axis=0).sort_values(ascending=False).head(50)
```

```python
sudet = cadaster_nn[cadaster_nn['parcelOwnerText'].str.contains(r"suddet|sudet|Suddet", case=False, na = False)]
```

```python
# we get he ids of which people contain suddetto 
suddeto_ids = sudet['id'].unique()

```

```python
for s in suddeto_ids:
    cadaster_nn.loc[cadaster_nn.index == s,'parcelOwnerText'] = cadaster_nn.loc[cadaster_nn.index == s-1,'parcelOwnerText']
```

```python
# we'll first develop the features we want
## first i want to find out when are there NAs? 
## we'll convert all the nan with None
cadaster_nn = cadaster_nn.where(pd.notnull(cadaster_nn), None)
```
